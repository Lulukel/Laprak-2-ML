```python
#IMPORT LIBRARY
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('botak.csv')
```

```python
df
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df["umur"], bins=20, kde=True)
plt.title("Distribusi Umur")
plt.xlabel("Umur")
plt.ylabel("Jumlah")
plt.show()
```

```python
proporsi_jenis_kelamin = df["jenis_kelamin"].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(proporsi_jenis_kelamin, labels=proporsi_jenis_kelamin.index, autopct="%1.1f%%")
plt.title("Distribusi Proporsi Jenis Kelamin")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x="pendidikan", data=df)
plt.title("Jumlah Orang Berdasarkan Pendidikan")
plt.xlabel("Pendidikan")
plt.ylabel("Jumlah")
plt.show()
```
